```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
image = cv2.imread('sunflower.jpg')
plt.figure(figsize=(8,8))
plt.axis("off")
plt.title("original image")
plt.imshow(image)
```

```python
grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.title("original image")
plt.imshow(grey_image)
```

```python
cv2.imwrite('sketch_grey_image.png',grey_image)
```

```python
invert = cv2.bitwise_not(grey_image)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.title("original image")
plt.imshow(invert)
```

```python
cv2.imwrite('sketch_inverted_image.png',invert)
```

```python
blur=cv2.GaussianBlur(invert,(21,21),0)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.title("original image")
plt.imshow(blur)
```

```python
cv2.imwrite('sketch_blur_image.png',blur)
```

```python
inverted_blur=cv2.bitwise_not(blur)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.title("original image")
plt.imshow(inverted_blur)
```

```python
cv2.imwrite('sketch_inverted_blur_image.png',inverted_blur)
```

```python
sketch=cv2.divide(grey_image,inverted_blur,scale=256.0)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.title("original image")
plt.imshow(sketch)
```

```python
cv2.imwrite('sketch.png',sketch)
```
